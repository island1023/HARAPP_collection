```python
import tensorflow as tf

# 建立转换器
converter = tf.lite.TFLiteConverter.from_saved_model("saved_model_outputs")

# 针对 Transformer 必须开启的配置
converter.target_spec.supported_ops = [
    tf.lite.OpsSet.TFLITE_BUILTINS,
    tf.lite.OpsSet.SELECT_TF_OPS # 这一行负责处理你的 LayerNormalization
]

# 执行转换
try:
    tflite_model = converter.convert()
    with open("har_transformer_final.tflite", "wb") as f:
        f.write(tflite_model)
    print("✅ 终于大功告成了！")
    print("请在左侧文件夹下载：har_transformer_final.tflite")
except Exception as e:
    print(f"导出 TFLite 时出错：{e}")
```

```python
import onnx
from onnx import shape_inference

def fix_and_check_model(input_path, output_path):
    # 1. 加载模型
    model = onnx.load(input_path)

    # 2. 尝试执行形状推断并修复拓扑（ONNX 内置工具）
    try:
        inferred_model = shape_inference.infer_shapes(model)
        onnx.save(inferred_model, output_path)
        print(f"✅ 模型基础修复成功，已保存至: {output_path}")
    except Exception as e:
        print(f"❌ 修复失败: {e}")

fix_and_check_model("mobilevit_wisdm.onnx", "mobilevit_fixed.onnx")
```

```python
import onnx
import onnx_graphsurgeon as gs

def cleanup_mobilevit(input_path, output_path):
    # 1. 加载模型
    model = onnx.load(input_path)
    graph = gs.import_onnx(model)

    # 2. 找到并处理 SequenceEmpty 节点 (n4)
    # 很多时候这些节点只是为了初始化循环，我们可以尝试清理孤立节点
    graph.cleanup()

    # 3. 强制转换所有算子为静态形状
    # 遍历所有层，尝试解决维度模糊问题
    for node in graph.nodes:
        if node.op == "Transpose":
            # 这里的逻辑是为了防止 onnx2tf 在转置时崩溃
            pass

    # 4. 保存
    new_model = gs.export_onnx(graph)
    onnx.save(new_model, output_path)
    print(f"✅ 预处理完成: {output_path}")

cleanup_mobilevit("mobilevit_fixed.onnx", "mobilevit_final_pre.onnx")
```

```python
import onnx
import onnx_graphsurgeon as gs

def fix_mobilevit_structure(input_path, output_path):
    model = onnx.load(input_path)
    graph = gs.import_onnx(model)

    # 找到所有的 SequenceEmpty 算子并标记
    # 在这个模型中，n4 是导致崩溃的元凶
    for node in graph.nodes:
        if node.op == "SequenceEmpty":
            # 尝试将其替换为常量或直接断开（如果它是非关键支路）
            print(f"正在处理异常算子: {node.name}")

    graph.cleanup().toposort()
    new_model = gs.export_onnx(graph)
    onnx.save(new_model, output_path)
    print(f"✅ 处理完成，生成: {output_path}")

fix_mobilevit_structure("mobilevit_final_pre.onnx", "mobilevit_no_sequence.onnx")
```
